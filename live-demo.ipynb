The Hamiltonian is standard


$$H =  \frac{\Delta E}{2} (\sigma_{zD_2} + \sigma_{zPd}) + \hbar\omega\left(a^{\dagger}a +\frac{1}{2}\right) + U\left( a^{\dagger} + a \right)(\sigma_{+D_2} +\sigma_{-D_2}) + U\left( a^{\dagger} + a \right)(\sigma_{+Pd} +\sigma_{-Pd})$$

```python
D2, Pd, vibrations, interaction_D2, interaction_Pd, state_list = make_operators_AB(max_bosons=1, num_TLS_A=1, num_TLS_B=1)
```

Now we construct the Hamiltonian

```python
H = 10.5*D2 + 10.5*Pd + 1*vibrations + 0.01*interaction_D2 + 0.01*interaction_Pd
```

Next we simulate

```python
psi0 = basis([2,2,2],[0,0,1])
times = np.linspace(0.0, 1000000, 10000)
P, psi, *_ = simulate(H, psi0, times)
```

And finally we visualise

```python
bra_labels, ket_labels = make_braket_labels(state_list)
plot_sim(times, P, ket_labels)
```
